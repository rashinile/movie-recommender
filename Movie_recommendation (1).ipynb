```python
import pandas as pd
movies = pd.read_csv("/content/tmdb_5000_movies.csv")
movies.shape
```

```python
movies.columns
movies.head()
```

```python
#select required columns
movies = movies[['title','overview','genres']]
movies.isnull().sum()
```

```python
movies['overview']=movies['overview'].fillna(' ')
movies.isnull().sum()
```

```python
import ast

def extract_genres(genres):
  genre_list = []
  for genre in ast.literal_eval(genres):
    genre_list.append(genre['name'])
  return " ".join(genre_list)

# Apply function to genres column
movies['genres'] = movies['genres'].apply(extract_genres)

# Check result
movies.head()
```

```python
# Create tags feature by combining overview and genres
movies['tags'] = movies['overview'] + " " + movies['genres']
movies.head()

#keep only necessary cols
movies =  movies[['title','tags']]
movies.head()
```

```python
#Data preporcessing
#convert tags into lower case
movies['tags']= movies['tags'].apply(lambda x:x.lower())
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#initialize vector
cv = CountVectorizer(max_features=5000 ,stop_words='english')
#convet tags to vectors
vectors = cv.fit_transform(movies['tags']).toarray()
vectors.shape
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#calculate cosine similarity
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
#create function to recommend movies
def recommend(movie):
  #get index of the movie
  movie_index = movies[movies['title']==movie].index[0]

  #get similarity score
  distances = similarity[movie_index]

  #sort movies based on similarity scores
  movies_list = sorted(
      list(enumerate(distances)),
      reverse=True,
      key=lambda x:x[1]
  )
  #return top 5 similar movies (excluding itself)
  recommend_movies = []
  for i in movies_list[1:6]:
    recommend_movies.append(movies.iloc[i[0]].title)
  return recommend_movies
```

```python
recommend('Avatar')
```

```python
import pickle

pickle.dump(movies, open('movies.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
```
